Non-spatial (one grid cell) run.

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import lsst.geom as geom
from astropy.visualization import ZScaleInterval
from lsst.daf.butler import Butler
```

```python
butler = Butler("/datasets/hsc/gen3repo/rc2w50_ssw02", collections='u/gkovacs/DM-29522', writeable=False)
```

```python
diffExp = butler.get("deepDiff_differenceExp", visit=1166, detector=10, band='z', instrument='HSC')
```

```python
# Show the image with the centroids of the selected sources
zScaler = ZScaleInterval()
vmin, vmax = zScaler.get_limits(diffExp.image.array)
bbox = diffExp.getBBox()

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
cb = ax.imshow(diffExp.image.array, origin='lower', interpolation='none', vmin=vmin, vmax=vmax,
              extent=[bbox.getMinX()-0.5, bbox.getMaxX()+0.5, bbox.getMinY()-0.5, bbox.getMaxY()+0.5])
ax = fig.add_axes((0.04,0.01,0.02,0.92))
fig.colorbar(cb, cax=ax, label='zscale')
```
